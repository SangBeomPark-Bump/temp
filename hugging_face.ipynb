LLM을 통과하면 대체로 마크다운 문법으로 온다.

---

LLM은 "대형 언어 모델" (Large Language Model)의 약자로, 인공지능(AI) 분야에서 자연어 처리를 위한 모델을 의미합니다. 이러한 모델은 대량의 텍스트 데이터에서 학습하여 인간과 유사한 방식으로 언어를 이해하고 생성할 수 있도록 설계되었습니다.

### LLM의 특징:
1. **사전 학습과 미세 조정**: LLM은 일반적으로 두 단계로 학습됩니다. 처음에는 대규모 텍스트 코퍼스를 사용해 사전 학습을 진행한 후, 특정 작업에 맞게 미세 조정(fine-tuning)합니다.

2. **Transformer 아키텍처**: 대부분의 LLM은 Transformer 아키텍처를 기반으로 합니다. 이 구조는 병렬 처리와 장기 의존성 학습에 강점을 가지고 있어 효과적인 언어 모델링에 적합합니다.

3. **거대한 파라미터 수**: LLM은 수억 또는 수십억 개의 매개변수(parameter)를 가질 수 있어, 더 복잡한 패턴을 학습하고 표현할 수 있습니다.

4. **응용 가능성**: LLM은 번역, 요약, 질문 응답, 대화 생성 등 다양한 자연어 처리 작업에 활용될 수 있습니다.

### 예시
- **GPT (Generative Pre-trained Transformer)**: OpenAI에서 개발한 모델로 다양한 버전이 있으며, 자연어 생성에 뛰어난 성능을 보입니다.
- **BERT (Bidirectional Encoder Representations from Transformers)**: Google에서 개발한 모델로 문맥의 양쪽에서 정보를 고려하여 이해하는 데 강점을 가집니다.

LLM은 인공지능 분야에서 큰 혁신을 가져왔으며, 여러 산업에서 채택되고 있습니다. 다만, 이들이 가진 윤리적, 사회적 문제도 함께 논의되고 있습니다.

- Meta에서 개발한 오픈소스 LLM
- LLama 3.1 8B, Llama 3.2 3B, 1B모델 등이 코랩에서 간단히 돌릴수 있는 최신 모델

```python
from langchain.llms import HuggingFacePipeline
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

# 사용할 LLaMA 모델 선택 (로컬에서 실행할 모델)
model_name = "meta-llama/Llama-3.2-3B-Instruct"  # 또는 "meta-llama/Llama-3.2-3B-Instruct"

# 토크나이저 및 모델 로드 (자동으로 GPU 사용)
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto")


```

```python
# Transformers 파이프라인 생성
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=256,  # 응답 최대 길이
    do_sample=True,       # 샘플링 활성화
    temperature=0.7,      # 창의적인 답변 조절
    repetition_penalty=1.03,  # 반복 단어 방지
)

# LangChain과 연동
llm = HuggingFacePipeline(pipeline=pipe)

# 실행
response = llm.invoke("LLM이란 무엇인가?")
print(response)
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

```python
model_name = "meta-llama/Llama-3.2-3B-Instruct" #### Llama 3.2 3B (30억개 파라미터) - Instruct (채팅용으로 Fine-tuning됨)
```

```python
from huggingface_hub import login
hf_token = "__" ### 토큰 입력
login(token=hf_token)
```

```python
model =  AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto", ### Cuda면 gpu, mps 등이면 mps, 안되면 cpu
    torch_dtype=torch.bfloat16, #### colab은 bfloat16이 지원됨. PC 그래픽카드의 경우 되는지 확인해봐야함.
    )
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
messages = [
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "Hello, how are you?"},
] ### chat_template의 template.
### List에 담겨있는건 예측이나 Train을 할때 batch_size만큼 train됐으므로 차원 맞춰주기용

inputs = tokenizer.apply_chat_template(
    messages,
    return_tensors="pt"
    ).to("cuda") ### GPU가 안되는 환경에선 "cpu"로
### 토크나이저를 거쳐 숫자로

outputs = model.generate(input_ids = inputs, max_new_tokens=256, temperature = 0.1) ### 최대 토큰수(모델이 일정량이상 토큰을 뱉으면 강제종료), temperature는 후에 설명

### output역시 숫자

generated_text = tokenizer.decode(
    outputs[0], ### batch에 담겨져있다는 가정때문에 [0]추가됨
    skip_special_tokens=True ### 스페셜 토큰들 안보기
)
### tokenizer.decode로 ouput으로 나온 숫자 변경

generated_text_raw = tokenizer.decode(outputs[0], skip_special_tokens=False) ### 스페셜 토큰들 보기
```

```python
print(generated_text)
### generated_text는 내가 물어본것 뿐만아니라 그냥 전부 다 나옴
### 내가 물어본부분만 알고싶으면 따로 인덱싱해주면 됨.
```

```python
print(generated_text_raw)
### 토크나이저가 바라보는 text.
### 뭐가 많은데 이건 나중에 알아보겠음
```

```python
from transformers import TextStreamer
text_streamer = TextStreamer(tokenizer, skip_prompt = True)
### streaming 모드
### 글자를 한글자 한글자 생성해내는 LLM의 특성상 스트리밍 모드로 실행하는게 기다리는데 덜지루함

_ = model.generate(input_ids = inputs, streamer = text_streamer, max_new_tokens = 128,
                   use_cache = True, temperature = 0.1)
```

### 모델의 output
- 기본 개념: 모델을 마지막으로 통과한 텐서는 tokenizer 크기 차원의 tensor가 됨.  
이때 softmax는 계산상 불리하기도하고, 그냥 logits나온 순서만 갖다써도 되므로 softmax를 따로 쓰지는 않음.   
다음 토큰을 예측할때 logits끼리 크기의 비율로 다음 토큰이 결정되는 방식.  
만약 a토큰 0.5, b토큰 0.3, c토큰 0.2면 각각 50%, 30%, 20%확률로 결정됨.  
단, logits끼리 합쳐서 1이 될 필요는 없음.
- temperature : temperature는 logits끼리 크기를 비슷하게 조정해줌. 커질수록 logits끼리 비슷해짐.  
위의 예시의 경우temperature가 커지면 0.37, 0. 33, 0.30 이런식으로 확률 조정이 들어간다는 뜻.

- min_p : 최소확률 조정  
아까의 예시의 경우 min_p를 0.25로 하면 c토큰은 logits를 0으로 치환해버림

- min_k : 최소 갯수
예를들어 min_k가 2면 가장 높은 logits의 토큰 2개만 사용해서 추론한다는 뜻

```python

from unsloth import FastLanguageModel
import torch
max_seq_length = 2048 # Choose any! We auto support RoPE Scaling internally!
dtype = None # None for auto detection. Float16 for Tesla T4, V100, Bfloat16 for Ampere+
load_in_4bit = True # Use 4bit quantization to reduce memory usage. Can be False.

model_name = "meta-llama/Llama-3.1-8B-Instruct"  # 예: "mistralai/Mistral-7B-v0.1"

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = model_name, # or choose "unsloth/Llama-3.2-1B-Instruct"
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
    # token = "hf_...", # use one if using gated models like meta-llama/Llama-2-7b-hf
)
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python

FastLanguageModel.for_inference(model) # Enable native 2x faster inference

messages = [
    {"role": "user", "content": "Continue the fibonnaci sequence: 1, 1, 2, 3, 5, 8,"},
]
inputs = tokenizer.apply_chat_template(
    messages,
    tokenize = True,
    add_generation_prompt = True, # Must add for generation
    return_tensors = "pt",
).to("cuda")

from transformers import TextStreamer
text_streamer = TextStreamer(tokenizer, skip_prompt = True)

_ = model.generate(input_ids = inputs, streamer = text_streamer, max_new_tokens = 2048,
                   use_cache = True, temperature = 0.1, min_p = 0.1)
```
